```
import pandas as pd
df1 = pd.read_csv("User.csv")
df2 = pd.read_csv("Period.csv")
df3 = pd.read_csv("Symptom.csv")
```

```
df1.head()
```

```
df = df1.iloc[:,2:]
df2.head()
```

```
main = pd.concat([df, df2], axis=1, join='inner')
```

```
main.head()
```

```
main = pd.concat([main, df3], axis=1, join='inner')
```

```
main.head()
```

```
main = main.drop(['id','start_date','end_date','User_id','id','user_id','date'],axis=1)
```

```
X = main.iloc[:,:2]
```

```
X.head()
```

```
Y = main.iloc[:,2:]
```

```
Y.head()
```

```
import tensorflow as tf
model = tf.keras.models.Sequential([
             tf.keras.layers.Dense(1000,activation = tf.nn.relu),
             tf.keras.layers.Dense(500,activation = tf.nn.relu),
             tf.keras.layers.Dense(9,activation = tf.nn.relu)
])
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
X_train = X_train.values
X_test = X_test.values
Y_test = Y_test.values
Y_train = Y_train.values
```

```
model.compile(optimizer = 'Adam',loss = 'categorical_crossentropy',metrics=['accuracy'])
```

```
model.fit(X_train,Y_train,epochs=10000)
```

```
pred = model.predict(X_test)
```
